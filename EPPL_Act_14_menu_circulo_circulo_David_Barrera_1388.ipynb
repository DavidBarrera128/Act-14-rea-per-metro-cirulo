redaccion: El objetivo de este proyecto es desarrollar un programa interactivo en lenguaje Python, dirigido a estudiantes de nivel preparatoria, que permita calcular de manera práctica tres propiedades geométricas fundamentales: el área de un círculo, el perímetro (longitud de la circunferencia) y el volumen de una esfera. El programa debe presentar un menú de opciones, ejecutar los cálculos mediante funciones independientes, mostrar las fórmulas aplicadas en cada caso y mantener una ejecución continua hasta que el usuario decida salir.

![texto del vínculo](https://raw.githubusercontent.com/DavidBarrera128/Mi-imagenes-de-EPPL-Unidad-2/refs/heads/main/circulo.PNG)

```python
import math

def areacirculo():
    """Calcula y muestra el área de un círculo."""
    print("\n📐 Fórmula utilizada: Área = π * r²")
    radio = float(input("🔵 Ingresa el radio del círculo: "))
    area = math.pi * (radio ** 2)
    print(f"✅ El área del círculo es: {area:.2f} unidades cuadradas\n")

def perimetrocirculo():
    """Calcula y muestra el perímetro (circunferencia) de un círculo."""
    print("\n📏 Fórmula utilizada: Perímetro = 2 * π * r")
    radio = float(input("🔵 Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro del círculo es: {perimetro:.2f} unidades lineales\n")

def volumenesfera():
    """Calcula y muestra el volumen de una esfera."""
    print("\n🌐 Fórmula utilizada: Volumen = (4/3) * π * r³")
    radio = float(input("🔵 Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"✅ El volumen de la esfera es: {volumen:.2f} unidades cúbicas\n")

def menu():
    """Menú interactivo que controla el flujo del programa."""
    while True:
        print("✨ " + "="*35)
        print("   🧮 MENÚ DE CÁLCULOS GEOMÉTRICOS 🧮")
        print("✨ " + "="*35)
        print("1️⃣  Calcular área del círculo")
        print("2️⃣  Calcular perímetro del círculo")
        print("3️⃣  Calcular volumen de la esfera")
        print("4️⃣  Salir")
        print("✨ " + "="*35)

        opcion = input("👉 Selecciona una opción (1-4): ")

        if opcion == '1':
            areacirculo()
        elif opcion == '2':
            perimetrocirculo()
        elif opcion == '3':
            volumenesfera()
        elif opcion == '4':
            print("\n👋 ¡Gracias por usar el programa! Hasta pronto. 🚀")
            break
        else:
            print("⚠️ Opción no válida. Por favor, elige entre 1 y 4.\n")

# Punto de entrada principal del programa
if __name__ == "__main__":
    menu()
```
